## WUE_metrics

```python
%%R
# Create dummy dataset with relevant variables
set.seed(123)
n <- 100  # number of data points

dummy_data <- data.frame(
  GPP = runif(n, 5, 20),         # umol CO2 m-2 s-1
  NEE = runif(n, -10, 10),       # umol CO2 m-2 s-1
  LE = runif(n, 50, 200),        # W m-2
  VPD = runif(n, 0.5, 2.5),      # kPa
  Tair = runif(n, 10, 30)        # deg C
)

write.csv(dummy_data, file = "dummy_data.csv", row.names = FALSE)

print(head(dummy_data))

print(WUE.metrics(dummy_data))
```
